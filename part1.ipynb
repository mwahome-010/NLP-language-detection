```python
#Cleanup transcribed audio dataset
```

```python
import csv

input_file = 'digital_green_recordings.csv'
output_file = 'cleaned_digital_green_recordings.csv'

with open(input_file, mode='r', encoding='utf-8') as infile:
    reader = csv.reader(infile)
    with open(output_file, mode='w', encoding='utf-8', newline='') as outfile:
        writer = csv.writer(outfile)
        for row in reader:
            if row: # Skip empty lines
                # Split the first column at '$' and keep the last part
                row[0] = row[0].split('$')[-1]
                writer.writerow(row)
```

```python
#Combine all CSVs
```

```python
import pandas as pd
import glob

files = glob.glob("clean_data/*.csv")

dfs = []

for file in files:
    # Due to mixed datatypes, forcing all to be string
    df = pd.read_csv(file, dtype=str)

    df.rename(columns={df.columns[0]: "Text"}, inplace=True)

    dfs.append(df)

master_df = pd.concat(dfs, ignore_index=True)

master_df.to_csv("combined_sentences.csv", index=False)
```
